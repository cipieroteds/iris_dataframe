```python
import sklearn
import numpy as np
import pandas as pd
```

```python
from sklearn import datasets
```

```python
iris =datasets.load_iris()
iris
```

```python
iris.data
```

```python
iris.target
```

```python
dados = iris.data
dados
```

```python
dados = pd.DataFrame(dados)
dados
```

```python
classes = iris.target
classes
```

```python
classes = pd.DataFrame(classes)
classes
```

```python
df_iris = pd.concat([dados,classes], axis=1)  #axis=1 concatenar os dois df com colunas
df_iris
```

```python
colnames = iris.feature_names
colnames
```

```python
type(colnames)
```

```python
colnames.append('classification')
colnames
```

```python
df_iris.columns = colnames
df_iris
```

```python
nomes = iris.target_names
nomes
```

```python
df_iris['classification'] = df_iris['classification'].replace({0:nomes[0],
                                                               1:nomes [1],
                                                               2:nomes[2]})
df_iris
```
